```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = 'D:\\Github\\2nd-ML100Days\\'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
unique_house_type = app_train[~app_train['HOUSETYPE_MODE'].isnull()]['HOUSETYPE_MODE'].unique()
unique_house_type
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train[~app_train['HOUSETYPE_MODE'].isnull()]['HOUSETYPE_MODE'].unique()
print(unique_house_type[0])

nrows = len(unique_house_type)
ncols = nrows // 2 #//取整數 

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1) #subplot(列,欄,顯示位子)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i],'AMT_CREDIT'].hist(rwidth = 0.8)
    #loc[ 這裡要找app_train['HOUSETYPE_MODE']等於unique_house_type裡面的值 ,欄]  
    #unique_house_type有三個值['block of flats', 'terraced house', 'specific housing']
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
